## **Laços de Repetição**

Laços de repetição servem para executar o mesmo código várias vezes, sem precisar copiar e colar.

### **Laço for**

O `for` é usado quando sabemos quantas vezes queremos repetir algo.

```python
for i in range(5):
    print(i)
```

* **for** → palavra reservada do Python que inicia um laço de repetição
* **i** → é uma variável criada pelo for
  * ela guarda o número da vez
  * pode se chamar i, numero, contador, etc.
* **in** → significa “dentro de”
* **range(5)** → cria uma sequência de números:
  * começa em 0
  * termina em 4
* **:** → indica que o bloco do for começa aqui
* Tudo que está indentado (com espaço à esquerda) roda dentro do for
* **print(i)** → mostra o valor atual de i

### **Pratique**

Mude o número 5 para 10 e observe o resultado.

```python
# escreva seu código aqui
for i in range(10):
    print(i)
```

### **`for` com início e fim definidos**
Exemplo:

```python
for i in range(1, 6):
    print(i)
```

Explicação:

O `range(1, 6):` começa em 1 e termina em 5 (o 6 não entra na contagem)

### **Pratique**

Faça um for que conte de 1 até 10.

```python
# escreva seu código aqui
for i in range(1, 11):
    print(i)
```

### **Laço while**

O `while` repete o código enquanto uma condição for verdadeira.

Exemplo:

```python
contador = 0 # Criamos uma variável chamada contador e ela começa valendo zero

while contador < 5: # while significa “enquanto”. O contador < 5 é uma condição. Enquanto essa condição for verdadeira, o código continua. Os ":" indica o início do bloco
    print(contador) # Mostra o valor atual do contador
    contador += 1 # Soma 1 ao valor atual e isso evita o loop infinito
```

⚠️ **Atenção:** Loop infinito

Se esquecermos de aumentar o contador, o while nunca para.

**Exemplo errado:**

```python
contador = 0

while contador < 5:
    print(contador)
```

### **Pratique**

Crie um while que conte de 1 até 10.

```python
# escreva seu código aqui
contador = 1
while contador <= 10 :
    print(contador)
    contador += 1
```

### **Lâmpada com ligar e desligar**

*Relembrando a ideia*

* **Classe** → modelo
* **Objeto** → coisa criada
* **Funções** → ações que a coisa pode fazer

Vamos criar uma classe com funções!

```python
class Lampada: # Cria o modelo da lâmpada
    def __init__(self): # Toda lâmpada começa desligada
        self.ligada = False

    def ligar(self): # Função que liga a lâmpada
        self.ligada = True # Muda o estado para ligada

    def desligar(self): # Função que desliga a lâmpada
        self.ligada = False # Muda o estado para desligada
```

### **Criando e usando a lâmpada**

```python
lampada = Lampada()

lampada.ligar()   # a acao vem da funcao
print(lampada.ligada)  #imprimindo o atributo

lampada.desligar()
print(lampada.ligada)
```

### **Pratique**
1. Crie uma lâmpada
2. Ligue
3. Desligue
4. Veja o valor mudar

```python
# escreva seu código aqui

class Lampada_eu:   #objeto , o atributo com init , funcao positiva , funcao negativa

  def __init__(self):
      self.ligada = False

  def ligar(self):
    self.ligada = True

  def desligar(self):
    self.ligada = False

lampada_exibida_eu = Lampada_eu()
lampada_exibida_eu.ligar()
print(lampada_exibida_eu.ligada)
lampada_exibida_eu.desligar()
print(lampada_exibida_eu.ligada)
```

## **Como os programadores podem usar classes nos sistemas das empresas?**

Eles podem criar um sistema para o pedido de uma loja

```python
class Pedido:
    def __init__(self):
        self.itens = []

    def adicionar_item(self, item):
        self.itens.append(item)

    def mostrar_itens(self):
        for item in self.itens:
            print(item)

pedido = Pedido()
pedido.adicionar_item("Camisa")
pedido.adicionar_item("Tênis")
pedido.mostrar_itens()
```

Podem criar um sistema para verificar se os alunos passaram ou não de ano de acordo com a nota

```python
alunos = []

while True:
    nome = input("Nome do aluno (ou sair): ")
    if nome == "sair":
        break

    nota = float(input("Nota: "))
    alunos.append({"nome": nome, "nota": nota})

for aluno in alunos:
    if aluno["nota"] >= 7:
        print(aluno["nome"], "aprovado")
    else:
        print(aluno["nome"], "reprovado")
```
