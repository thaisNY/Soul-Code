## **O que é uma coleção**

Em Python, uma **coleção** é uma estrutura usada para **guardar vários valores dentro de uma única variável**, em vez de criar uma variável para cada valor.

Ela serve para **organizar, agrupar e trabalhar com conjuntos de dados relacionados**.

### Pensando de forma simples

Ao invés de fazer isso:

````python
a = 1
b = 2
c = 3
````

Você usa uma **coleção**:

````python
numeros = [1, 2, 3]
````

Uma única variável passa a representar **vários valores**.

### O que uma coleção permite fazer

Coleções permitem:

* armazenar **vários elementos**
* percorrer os dados com `for`
* verificar se algo existe (`in`)
* aplicar funções como `len()`, `max()`, `min()`
* organizar dados de formas diferentes, dependendo do tipo

### Tipos comuns de coleções em Python

Cada coleção tem um comportamento específico:

* **list** → ordenada, mutável, aceita repetição
* **tuple** → ordenada, não pode ser alterada
* **set** → não ordenada, não aceita repetição
* **dict** → pares de chave e valor

Todas são coleções porque **guardam vários dados juntos**, mas cada uma resolve um tipo de problema diferente.

### **Métodos de Tuplas**

* **`count(valor)`**
  Retorna **quantas vezes** um determinado valor aparece dentro da tupla.
  É útil para verificar repetições, já que tuplas podem conter valores duplicados.

* **`index(valor)`**
  Retorna o **índice da primeira ocorrência** do valor informado na tupla.
  Se o valor não existir, ocorre um erro (`ValueError`).



````python
t = (1, 2, 2, 3)

print(t.count(2))
# Saída: 2
# O método count conta quantas vezes o valor 2 aparece na tupla. Como o número 2 aparece duas vezes, o resultado é 2.

print(t.index(3))
# Saída: 3
# O método index retorna a posição (índice) da PRIMEIRA ocorrência do valor informado. O número 3 está na posição 3 da tupla.
````

### **Tuplas como Chaves de Dicionário**

Um **dicionário** funciona como um sistema de **localização rápida**:
dada uma chave, o Python precisa encontrar o valor **sempre no mesmo lugar**.

Para isso, o Python cria uma espécie de **assinatura interna** da chave no momento em que ela é usada.

---

### O que o Python espera de uma chave?

* Que ela **não mude de forma**
* Que ela **represente sempre a mesma coisa**
* Que o Python consiga **reconhecê-la exatamente igual toda vez**

Quando você usa uma tupla como chave, isso funciona bem porque:

````python
(2, 3)  # sempre representa exatamente esse par
````

````python
(2, 3) # nunca vira outra coisa.
````

### O problema das listas como chave

Uma lista **pode mudar de conteúdo**, mesmo sendo “a mesma variável”:

````python
posicao = [2, 3]
# depois…
posicao[0] = 9
````

Agora aquela “posição” **já não representa mais o mesmo ponto**.

Para o dicionário, isso é um problema sério: ele não sabe mais onde o valor deveria estar.

Por isso, o Python **nem permite** listas como chave.

Tuplas funcionam como chaves porque **representam algo estável e confiável**, enquanto listas podem mudar e **quebrar a lógica de localização do dicionário**.

Essa regra existe para garantir que o dicionário **sempre encontre o valor certo**, rápido e sem ambiguidade.

````python
sobre_mim = {
    (0, 0): "Você está no começo, e tudo bem começar do zero",
    (1, 2): "Você já avançou, cada passo importa",
    (2, 3): "Seu progresso é real, continue acreditando em você"
}

print(sobre_mim)
````

### **Acessando chaves e valores em um dicionário com tuplas**
### Como acessar um valor usando a chave

Para acessar um valor no dicionário, você deve informar **a chave exata**, que neste caso é uma **tupla**:

````python
sobre_mim = {
    (0, 0): "Você está no começo, e tudo bem começar do zero",
    (1, 2): "Você já avançou, cada passo importa",
    (2, 3): "Seu progresso é real, continue acreditando em você"
}

print(sobre_mim[(1, 2)])
````

````python
sobre_mim = {
    (0, 0): "Você está no começo, e tudo bem começar do zero",
    (1, 2): "Você já avançou, cada passo importa",
    (2, 3): "Seu progresso é real, continue acreditando em você"
}

print(sobre_mim.values())
````

* Retorna todas as mensagens armazenadas no dicionário

### Como percorrer chaves e valores juntos

````python
sobre_mim = {
    (0, 0): "Você está no começo, e tudo bem começar do zero",
    (1, 2): "Você já avançou, cada passo importa",
    (2, 3): "Seu progresso é real, continue acreditando em você"
}

for posicao, mensagem in sobre_mim.items():
    print(posicao, "->", mensagem)
````

**O que acontece no loop:**

* `posicao` recebe a tupla (chave)
* `mensagem` recebe o texto associado a essa chave

## **enumerate**

### **O que é**

Permite percorrer uma sequência obtendo **índice + valor** ao mesmo tempo.

````python
desenhos_da_disney = ["Bela", "Tarzan", "Operação Big Hero"]

for indice, desenho in enumerate(desenhos_da_disney):
    print(indice, desenho)
````

O enumerate **não imprime automaticamente os elementos**

Ele cria um objeto iterável

O Python então mostra apenas onde esse objeto está na memória

Por isso aparece algo como: <enumerate at 0x7dcdc656bb50>

````python
desenhos_da_disney = ["Bela", "Tarzan", "Operação Big Hero"]
enumerate(desenhos_da_disney, start=1)
````

Isso é só a representação interna do objeto, não o conteúdo.

#### **Como ver os valores de verdade**

Você precisa **percorrer** ou **converter** esse objeto.

**Usando** `for`



````python
desenho_da_disney = "Operação Big Hero"

for indice, letra in enumerate(desenho_da_disney, start=1):
    print(indice, letra)
````

**Convertendo para lista**

````python
desenhos_da_disney = ["Bela", "Tarzan", "Operação Big Hero"]
print(list(enumerate(desenhos_da_disney, start=1)))
````

Forma **concisa e legível** de criar listas a partir de iteráveis.

````python
# Lista de livros
livros = ["Dom Casmurro", "O Cortiço", "Vidas Secas", "Capitães da Areia"]

# List comprehension: cria uma lista com o tamanho do nome de cada livro
tamanhos_dos_livros = [len(livro) for livro in livros]

print(tamanhos_dos_livros)
# Cada número corresponde à quantidade de caracteres do título do livro
````

### **Com condição**

````python
# Lista de livros
livros_carolina_maria_de_jesus = [
    "Quarto de Despejo",
    "Casa de Alvenaria",
    "Pedaços da Fome",
    "Provérbios",
    "Diário de Bitita"
]

# List comprehension com condição: seleciona apenas livros com mais de 12 caracteres no nome
livros_longos = [
    livro for livro in livros_carolina_maria_de_jesus
    if len(livro) < 12
]

print(livros_longos)
# Retorna apenas os títulos cujo nome é mais longo
````

##**arg** e **kwargs**

### **O que são**

* `*args` → múltiplos argumentos posicionais
* `**kwargs` → múltiplos argumentos nomeados

````python
# *args → permite receber vários valores posicionais
def soma_pontos(*args):
    total = sum(args)
    return total


# **kwargs → permite receber vários argumentos nomeados
def info_jogador(**kwargs):
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")


# usando *args
pontos = soma_pontos(10, 20, 15, 5)
print("Pontuação total:", pontos)

# usando **kwargs
info_jogador(
    nome="Lara",
    nivel=12,
    classe="Arqueira",
    vida=350
)
````

## **Funções `lambda`**

### **O que é**

`lambda` é uma forma **curta e simples** de criar funções em Python, usada quando a lógica é pequena e direta.

Pense assim:

> **`lambda` é uma função escrita em uma única linha.**

Ela funciona como uma função normal, mas:

* não usa `def`
* não usa `return`
* retorna o resultado automaticamente

* **antes dos `:`** → o que a função recebe
* **depois dos `:`** → o que a função devolve

````python
# Forma tradicional
def dobro(x):
    return x * 2

# Forma lambda
dobro = lambda x: x * 2

print(dobro(5))
````

As duas funções fazem exatamente a mesma coisa.

### **Por que usar `lambda`**

Use `lambda` quando:

* a função é **simples**
* será usada **poucas vezes**
* você quer deixar o código **mais direto**

Evite `lambda` quando:

* a lógica é grande
* há várias condições
* o código fica difícil de ler

````python
# Soma simples
soma = lambda a, b: a + b
print(soma(2, 3))  # 5

# Transformação de texto
maiusculo = lambda texto: texto.upper()
print(maiusculo("python"))  # PYTHON

# Verificação lógica
par = lambda n: n % 2 == 0
print(par(4))  # True
````

A lambda é usada apenas para definir **o que fazer com cada item**.

## **reduce**

### **O que é**

`reduce` é uma função que **pega vários valores e combina todos eles em um único resultado final**.

Ela aplica uma função **passo a passo**, acumulando o resultado.

Pense assim:

> **reduce vai “juntando” os elementos da lista até sobrar apenas um valor.**

### **Como funciona na prática**

1. Pega os dois primeiros valores
2. Aplica a função
3. Usa o resultado junto com o próximo valor
4. Repete até acabar a sequência


````python
from functools import reduce # Veremos sobre essa linha mais para a frente

# Lista de números
numeros = [2, 3, 4, 5]

# reduce:
# multiplica todos os valores da lista
resultado = reduce(lambda acumulador, atual: acumulador * atual, numeros)

print(resultado)
````

### **O que está acontecendo**

Passo a passo:

* 2 * 3 = 6
* 6 * 4 = 24
* 24 * 5 = 120

No final, sobra **um único valor**: `120`

````python
from functools import reduce

# Pontuações de um jogador
pontuacoes = [10, 15, 20, 5]

total = reduce(lambda total, pontos: total + pontos, pontuacoes)

print(total)
````

O `reduce` soma todas as pontuações em um único número.

## **Decorators**

### **O que são**

Decorators são **funções que envolvem outras funções**, adicionando comportamento **antes e/ou depois** da execução delas **sem modificar o código original**.

Pense assim:

> **Um decorator é como colocar uma “camada extra” em uma função.**

### **Por que decorators existem**

Eles servem para:

* evitar repetição de código
* adicionar funcionalidades comuns (log, validação, tempo de execução)
* deixar o código mais organizado e reutilizável



````python
# Função normal (sem decorator)
def jogar():
    print("O jogador iniciou a partida")

# Criando o decorator
def log_de_execucao(func):
    def funcao_interna():
        print("Iniciando ação...")
        func()
        print("Ação finalizada.")
    return funcao_interna

jogar = log_de_execucao(jogar)
jogar()
````

### Forma mais usada (com `@`)

````python
def log_de_execucao(func):
    def funcao_interna():
        print("Iniciando ação...")
        func()
        print("Ação finalizada.")
    return funcao_interna

@log_de_execucao #O `@log_de_execucao` é apenas uma forma mais bonita de escrever
def jogar():
    print("O jogador iniciou a partida")

jogar()
````

### O que está acontecendo por trás

1. Python lê a função `jogar`
2. Aplica o decorator `log_de_execucao`
3. Substitui `jogar` por `funcao_interna`
4. Quando você chama `jogar()`, quem roda é `funcao_interna()`

````python
def meu_decorator(func):
    def funcao_interna(nome):
        print("Olá!")
        func(nome)
        print("Até mais!")
    return funcao_interna

@meu_decorator
def saudacao(nome):
    print(f"Seja bem-vindo(a), {nome}")

saudacao("Ana")
````

### Regra importante para não quebrar código

Use `*args` e `**kwargs` na `funcao_interna`:

* Assim, o decorator funciona com **qualquer função**.

````python
def meu_decorator(func):
    def funcao_interna(*args, **kwargs):
        print("Antes")
        resultado = func(*args, **kwargs)
        print("Depois")
        return resultado
    return funcao_interna
````

##**Encapsulamento**

Encapsulamento é o conceito de **proteger os dados internos de um objeto**, controlando como eles podem ser acessados ou modificados.

Vamos pensar nos nossos jogos de vídeo game. Isso aparece para:

* impedir que a **vida do personagem seja alterada diretamente**
* garantir que regras do jogo sejam respeitadas (ex: não ter vida negativa)

Em Python, isso é feito por convenção:

* `_atributo` → uso interno
* `__atributo` → name mangling (mais protegido)

````python
class Personagem:
    def __init__(self, nome, vida):
        self.nome = nome
        self.__vida = vida  # atributo encapsulado

    def sofrer_dano(self, dano):
        if dano > 0:
            self.__vida -= dano
            if self.__vida < 0:
                self.__vida = 0

    def ver_vida(self):
        return self.__vida
````

O atributo `__vida` **não deve ser acessado diretamente fora da classe**, apenas por métodos controlados.

##**Herança**

Herança permite que uma classe **herde atributos e métodos de outra**, evitando repetição de código.

Em jogos, isso é muito comum:

* todos os personagens têm nome e vida
* cada tipo ataca de forma diferente

````python
class Personagem:
    def __init__(self, nome):
        self.nome = nome

    def atacar(self):
        print("O personagem ataca")

class Guerreiro(Personagem):
    def atacar(self):
        print("O guerreiro ataca com espada")

class Mago(Personagem):
    def atacar(self):
        print("O mago lança uma magia")

# Instanciando os personagens
heroi = Personagem("Genérico")
arthur = Guerreiro("Arthur")
merlin = Mago("Merlin")

# Chamando os métodos
heroi.atacar()   # Saída: O personagem ataca
arthur.atacar()  # Saída: O guerreiro ataca com espada
merlin.atacar()  # Saída: O mago lança uma magia
````

`Guerreiro` e `Mago` **herdam de `Personagem`** e sobrescrevem o método `atacar`.

##**Polimorfismo**

Polimorfismo significa que **objetos diferentes podem responder ao mesmo método de formas diferentes**.

Isso permite tratar todos os personagens de forma genérica.

````python
class Guerreiro:
    def atacar(self):
        return "Ataque físico"

class Mago:
    def atacar(self):
        return "Ataque mágico"

def executar_ataque(personagem):
    print(personagem.atacar())

# Criando os objetos
g1 = Guerreiro()
m1 = Mago()

# Usando a função genérica para mostrar o resultado
executar_ataque(g1)  # Saída: Ataque físico
executar_ataque(m1)  # Saída: Ataque mágico
````

A função funciona com **qualquer personagem** que possua o método `atacar`.

##**Métodos Mágicos**

Métodos mágicos (ou *dunder methods*) começam e terminam com `__`
Eles permitem que objetos se comportem como tipos nativos do Python.

Em jogos, isso ajuda a:

* exibir personagens de forma amigável
* comparar força, nível ou itens
* usar funções como `len()` diretamente

Exemplos comuns:

* `__init__` → construtor
* `__str__` → representação em texto
* `__eq__` → comparação
* `__len__` → tamanho

````python
class Item:
    def __init__(self, nome, poder):
        self.nome = nome
        self.poder = poder

    def __str__(self):
        return f"{self.nome} (Poder: {self.poder})"

    def __eq__(self, outro):
        return self.poder == outro.poder

# Criando itens
espada = Item("Espada de Ferro", 50)
cajado = Item("Cajado Arcano", 50)
adaga = Item("Adaga Curta", 20)

# Testando o __str__ (Representação em texto)
print(f"Item 1: {espada}")
print(f"Item 2: {cajado}")

# Testando o __eq__ (Comparação de igualdade baseada no poder)
print("---")
print(f"A Espada e o Cajado têm o mesmo poder? {espada == cajado}") # True (50 == 50)
print(f"A Espada e a Adaga têm o mesmo poder? {espada == adaga}")   # False (50 == 20)
````

##**Tratamento de Erros**

Tratar erros é como colocar uma rede de proteção.

No mundo da programação, dividimos os erros em três categorias principais. Entender a diferença entre eles ajuda muito na hora de "debugar" (consertar) seu código.

### 1. Erros de Sintaxe (Os Erros de Gramática)

Estes acontecem antes mesmo do programa rodar. É quando você escreve algo que o Python não entende, como esquecer dois pontos (`:`) ou fechar um parêntese errado.

* **Exemplo:** Escrever `if x = 10` em vez de `if x == 10`.
* **Como tratar:** O próprio editor (como o VS Code ou o Colab) geralmente sublinha de vermelho. Basta corrigir a escrita.


### 2. Erros de Exceção (Os Erros de Execução)

O código está escrito corretamente, mas algo impossível acontece enquanto ele está rodando. Por exemplo: tentar dividir um número por zero ou abrir um arquivo que não existe.

Para esses, usamos o bloco **`try / except`**. É a forma mais comum de "tratamento de erro".

### Como funciona na prática:

````python
try:
    numero = int(input("Digite um número: "))
    resultado = 10 / numero
    print(f"Resultado: {resultado}")
except ZeroDivisionError:
    print("Erro: Você não pode dividir por zero!")
except ValueError:
    print("Erro: Por favor, digite um número válido e não uma letra.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
finally:
    print("Fim da operação.") # Isso sempre será executado
````

### 3. Erros de Lógica (Os Erros "Silenciosos")

Esses são os mais perigosos. O programa roda perfeitamente, não exibe nenhuma mensagem de erro, mas o **resultado está errado**.

* **Exemplo:** Você quer calcular a média de duas notas, mas esquece os parênteses: `media = nota1 + nota2 / 2`. O Python vai dividir a `nota2` primeiro e depois somar, dando um valor absurdo.
* **Como tratar:**
  * **Testes Unitários:** Criar pequenos testes para verificar se as funções devolvem o que deveriam.
  * **Prints/Debug:** Colocar `print()` em partes estratégicas para ver o valor das variáveis.
  * **Assertions (`assert`):** Forçar o programa a parar se uma condição lógica for falsa.
